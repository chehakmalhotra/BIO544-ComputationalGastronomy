```python
import pandas as pd
import ast
from collections import Counter

data = pd.read_csv('/Users/chehakmalhotra/Desktop/untitled folder 2/ingredients_data.csv')
```

```python
data['Ingredients'] = data['Ingredients'].apply(ast.literal_eval) #to convert the string lists to actual lists
```

```python
all_ingredients = sum(data['Ingredients'], [])


ingredient_counts = Counter(all_ingredients)

for ingredient, count in ingredient_counts.items():
    print(f'{ingredient}: {count}')
```

```python
ingredient_counts_df = pd.DataFrame(ingredient_counts.items(), columns=['Ingredient', 'Frequency'])

print(ingredient_counts_df)
ingredient_counts_df.to_csv('ingredient_frequencies.csv')
```

```python
data['Recipe Size'] = data['Ingredients'].apply(len)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(data['Recipe Size'])
plt.title('Recipe Size Distribution')
plt.xlabel('Number of Ingredients')
plt.ylabel('Number of Recipes')


plt.savefig('recipe_size_distribution.png')


average_size = round(data['Recipe Size'].mean())


print(f'Average Recipe Size: {average_size}')
```

```python
sorted_sizes = sorted(data['Recipe Size'])
cumulative_counts = [i / len(sorted_sizes) for i in range(len(sorted_sizes))]


plt.figure(figsize=(10, 6))
plt.step(sorted_sizes, cumulative_counts)
plt.title('Cumulative Distribution of Recipe Sizes')
plt.xlabel('Number of Ingredients')
plt.ylabel('Cumulative Proportion')



plt.savefig('cumulative_distribution.png')

plt.show()
```
