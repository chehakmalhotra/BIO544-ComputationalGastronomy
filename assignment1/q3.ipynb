```python
import pandas as pd
import ast
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('/Users/chehakmalhotra/Desktop/untitled folder 2/ingredients_data.csv')
data['Ingredients'] = data['Ingredients'].apply(ast.literal_eval) #to convert the string lists to actual lists
data['Recipe Size'] = data['Ingredients'].apply(len)

size_counts = Counter(data['Recipe Size'])


sorted_sizes = sorted(size_counts.items(), key=lambda x: x[1], reverse=True) #sort des


sizes, frequencies = zip(*sorted_sizes)
ranks = range(1, len(sizes) + 1) #all ranks

total_recipes = len(data)
normalized_frequencies = [freq / total_recipes for freq in frequencies]


log_ranks = np.log(ranks)
log_normalized_frequencies = np.log(normalized_frequencies)

# Plot the log-log distribution
plt.figure(figsize=(10, 6))
plt.plot(log_ranks, log_normalized_frequencies)
plt.title('Log-Log Plot of Frequency-Rank Distribution of Recipe Sizes')
plt.xlabel('Log(Rank)')
plt.ylabel('Log(Normalized Frequency)')




plt.savefig('frequency_rank_distribution.svg')
plt.savefig('frequency_rank_distribution.png')

plt.show()
```
